```python
%%sql
SELECT * FROM sqlite_master
WHERE type='table';
```

```python
%%sql
--Checking column headers

SELECT * FROM facts
LIMIT 5
```

```python
%%sql

--Finding min and max

SELECT min(population), max(population), 
min(population_growth), max(population_growth)
FROM facts
```

```python
%%sql
--Finding smallest and largest poulations in dataset


SELECT * FROM facts
where population== (SELECT min(population) FROM facts)
or population== (SELECT max(population) FROM facts)
```

```python
%%sql

--Finding largest population size excluding 'World'

SELECT min(population), max(population), 
min(population_growth), max(population_growth)
FROM facts
WHERE name<>'World'

```

```python
%%sql
--Finding average population and area of dataset

SELECT avg(population), avg(area) FROM facts
```

```python
%%sql

/*Finding which countries have both above average populations
and below average area*/

SELECT * FROM facts
WHERE population> (SELECT avg(population) FROM facts) AND
area< (SELECT avg(area) FROM facts)
```

```python
%%sql
--Finding country with the largest population

SELECT * FROM facts
WHERE population== (SELECT max(population) FROM facts
                   WHERE name<> 'World')
```

```python
%%sql

--Finding countries with the highest water-to-land ratios

SELECT *,CAST(area_water AS FLOAT)/area_land as water_land_ratio, area_land/
area_water as land_water_ratio FROM facts
WHERE water_land_ratio<>'None'
ORDER BY water_land_ratio DESC
```

```python
%%sql
--Finding which countries have more water than land

SELECT *,CAST(area_water AS FLOAT)/area_land as water_land_ratio, 
CAST(area_land as FLOAT)/area_water as land_water_ratio FROM facts
WHERE water_land_ratio>1
ORDER BY water_land_ratio DESC

```

```python
%%sql
/*Finding which countries are adding the most people to their
populations, excluding 'World' and 'None' values*/

SELECT *,population*population_growth/100 as people_added
FROM facts
WHERE people_added<>'None' and name<>'World'
ORDER BY people_added DESC

```

```python
%%sql

/*Finding which countries have a higher death-rate than
birth-rate, and ordering by the ratio of deaths to births*/

SELECT *,death_rate/birth_rate as loss_ratio FROM facts
WHERE death_rate>birth_rate
ORDER BY loss_ratio DESC
```
